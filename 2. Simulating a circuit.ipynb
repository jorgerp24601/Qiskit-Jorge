```python
# We are going to be recreating the same experiment as in the past tutorial, so we will use the same commands:
import numpy as np
from qiskit import *
%matplotlib inline

#Now we create te qubits and the circuit
q = QuantumRegister(3, 'q')
circ = QuantumCircuit(q)

#And we apply the gates
circ.h(q[0])
circ.cx(q[0], q[1])
circ.cx(q[0], q[2])

#We will show the circuit too to keep track of what we are doing
circ.draw(output='mpl')
```

```python
# Now we want to simulate the circuit
# In order to do that we have to use Qiskit Aer
# Qiskit Aer is the Qiskit provider to simulate quantum circuits
# To call this provider we need to use (from qiskit import Aer)
# With the asterisk in the first command we dont need to use that
```

```python
# Now we need to select a backend
# The backend we need to use is the one stated bellow:

backend = BasicAer.get_backend('statevector_simulator')

# This backend will give us the quantum stave vector with 2^n dimensions with n the number of qubits
```

```python
# Now we want to to send the backend which work does it have to do
# To do that, we use the following command:

job = execute(circ, backend)

# This command will execute the circuit defined as circ in the defined backend
```

```python
# Now that the job is defined to extract the results we use the following command:

result = job.result()

#This will give us the results of the simulation
```

```python
# Now that we have the result, we have to extract the state that the simulation gives us
# In order to do that we run the following command:

outputstate = result.get_statevector(circ)
```

```python
# And now that we have obtained the output state we can print it via:

print(outputstate)
```

```python
# In order to plot the results of the state vector we have to call the right package

from qiskit.visualization import plot_state_city 

# This will import everything from the qiskit.visualization library
```

```python
# And finally, this following command will give us the representation of the state

plot_state_city(outputstate)

# This plots the density matrix both the real and the imaginary part of it
```

```python
# Another way of doing this calculus is by understanding each of the elements in the circuit as unitary operations
# In order to do that, we have to run a different backend

backend2 = BasicAer.get_backend('unitary_simulator')
```

```python
# Now that we have our new backend we need to execute a new job

job2 = execute(circ, backend2)
```

```python
# We get the new results

result2 = job2.result()
```

```python
# And we print them

print(result2.get_unitary(circ))
```
